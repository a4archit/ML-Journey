```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN, AgglomerativeClustering
```

```python
data = pd.read_csv(r'C:\\Users\\Archit\\Documents\\GitHub\\ML-Journey\\Datasets\\iris.csv')
```

```python
df = data.drop(columns=["Unnamed: 0","Id","Species"])
```

```python
df.sample(3)
```

```python
plt.title("Before clustering")
sns.scatterplot(data, x='PetalLengthCm',y='PetalWidthCm', hue='Species', palette='bright')
```

```python
db = DBSCAN(eps=0.5, min_samples=10)
cluster = db.fit_predict(df)
df['cluster'] = cluster
plt.title("After clustering")
sns.scatterplot(df, x='PetalLengthCm',y='PetalWidthCm', hue='cluster', palette='bright')
```

```python
db = AgglomerativeClustering(n_clusters=3)
cluster = db.fit_predict(df)
df['cluster'] = cluster
plt.title("After clustering")
sns.scatterplot(df, x='PetalLengthCm',y='PetalWidthCm', hue='cluster', palette='bright')
```
