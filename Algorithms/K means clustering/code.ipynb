```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

from sklearn.cluster import KMeans
```

```python
labelled_data = pd.read_csv(r"C:\\Users\\Archit\\Documents\\GitHub\\ML-Journey\Datasets\\iris.csv" ).drop(columns=['Unnamed: 0','Id'])
```

```python
plt.title("Before clustering")
sns.scatterplot(labelled_data, x='PetalLengthCm', y='PetalWidthCm', hue='Species', palette='bright')
```

```python
data = labelled_data.drop(columns=['Species'])
```

```python
data
```

```python
sns.scatterplot(data, x='PetalLengthCm', y='PetalWidthCm')
```

```python
kmc = KMeans(n_clusters=3)
clusters = kmc.fit_predict(data)
```

```python
data['clusters'] = clusters
```

```python
data
```

```python
def replace_numbers_by_class(num):
    if num == 0:
        return "Iris-virginica"
    elif num == 1:
        return "Iris-setosa"
    else:
        return "Iris-versicolor"
    
data['clusters'] = data['clusters'].apply(replace_numbers_by_class)
```

```python
plt.title("After clustering")
sns.scatterplot(data, x='PetalLengthCm',y='PetalWidthCm', hue='clusters', palette='bright')
```
