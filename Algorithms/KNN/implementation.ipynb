# KNN 

we do prediction using KNN (Regression)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, root_mean_squared_error, r2_score, confusion_matrix
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, StandardScaler
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
```

```python
df = pd.read_csv(r"/workspaces/ML-Journey/Datasets/Dataset for weight prediction with gender.csv")
```

```python
df
```

```python
X = df.drop(columns=['weight_kg'])
y = df[['weight_kg']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
scaler = StandardScaler()
X_train_trf = scaler.fit_transform(X_train)
X_test_trf = scaler.transform(X_test)
```

### Model without scaling

```python
knn = KNeighborsRegressor()
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
r2_score(y_test, y_pred)
```

### Model with scaling

```python
knn_s = KNeighborsRegressor()
knn.fit(X_train_trf, y_train)
y_pred = knn.predict(X_test_trf)
r2_score(y_test, y_pred)
```

### Optimising hyperparameters

```python
knn2 = KNeighborsRegressor(n_neighbors=10, weights='distance', algorithm='ball_tree', leaf_size=100)
knn2.fit(X_train_trf, y_train)
y_pred = knn2.predict(X_test_trf)
r2_score(y_test, y_pred)
```

### Using Grid Search CV

```python
param_grid = {
    'n_neighbors':[5,7,10,20],
    'weights':['distance','uniform'],
    'algorithm':['ball_tree','brute','kd_tree'],
    'leaf_size':[5,10,20,35,50],
    'n_jobs':[-1]
}

grid = GridSearchCV(estimator=knn, param_grid=param_grid, cv=5, scoring='neg_root_mean_squared_error')

grid.fit(X_train_trf, y_train)
```
