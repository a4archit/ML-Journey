# Multilinear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

from sklearn.linear_model import LinearRegression, Ridge
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score
```

```python
# taking data
X,y = make_regression(n_samples=10000, n_features=2, n_targets=1, n_informative=2)
```

```python
data = pd.DataFrame(X)
```

```python
data.columns = ['s1','s2']
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# model building
model = LinearRegression()
model.fit(X_train, y_train) # model understanding the data patterns
```

```python
x,y = model.coef_
```

```python
c = model.intercept_
```

```python
a,b = X_test[0]
```

```python
# Predicting manually after extracting values of general equation
# ax + by + c = 0
(a*x) + (b*y) + c
```

```python
y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
score_ = model.score(X_test, y_test)
```
