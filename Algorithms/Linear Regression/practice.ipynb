```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Linear Regression
- In this notebook we learn concepts about **Linear Regression** `Alggorithm`

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, accuracy_score
from math import sqrt
```

```python
# taking data

data = pd.read_csv(r"bodyPerformance.csv", usecols=['height_cm', 'weight_kg'])
```

```python
X = data[['height_cm']]
y = data[['weight_kg']]
```

```python
data.shape
```

```python
# train test splitting

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train) # model learnning the data points
```

```python
y_pred = model.predict(X_test)
```

```python
print(f"\
      MAE: {mean_absolute_error(y_test, y_pred)}\
      MSE: {sqrt(mean_squared_error(y_test, y_pred))}\
      R2 score: {r2_score(y_test, y_pred)}\
      ")
```

```python
# checking outliers

sns.boxplot(data['weight_kg'])
```

```python
sns.scatterplot(x=X_train['height_cm'], y=y_train['weight_kg'])
sns.scatterplot(x=X_test['height_cm'], y=y_test['weight_kg'])
plt.plot(X_test, y_pred, color='red')
```

```python
# getting the values of slope 

print(f"slope, m = {model.coef_[0,0]:.2f}")
```

```python
# getting the values of biases

model.intercept_
```
