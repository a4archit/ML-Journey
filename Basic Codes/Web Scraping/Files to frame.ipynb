```python


def to_csv_files(folder_name, file_name, lang):
    df = pd.DataFrame(columns=['name', 'username', 'location', 'repositories', 'followers', 'language','sorting type','description'])
    
    file_data2 = open(f"/workspaces/ML-Journey/Files to frame.ipynb", 'r', encoding='utf-8')
    content2 = file_data2.read()
    s = BeautifulSoup(content2, 'html')

    names2 = []
    username2 = []
    address2 = []
    repositories2 = []
    followers2 = []
    language2 = []
    sort_type2 = []
    description2 = []

    try:
        for box in s.find_all('div', class_="Box-sc-g0xbh4-0 iwUbcA"):
            box_data = box.find_all('span', class_="prc-Text-Text-0ima0")

            names2.append(box_data[0].text)
            username2.append(box_data[1].text)
            description2.append(box_data[3].text)
            repositories2.append(box_data[-3].text)
            followers2.append(box_data[-1].text)
            language2.append(lang)
            address2.append(box.find('li').text)
            
            sort_type = s.find_all('span', class_="Box-sc-g0xbh4-0 gUkoLg")[1].text.split(":")[-1].strip()
            sort_type2.append(sort_type)

    except Exception as e:
        print(f"Error in {file_name}:\n\t{e}")

    data = {
        'name':names2,
        'username':username2,
        'location':address2,
        'repositories':repositories2,
        'followers':followers2,
        'language':language2,
        'sorting type':sort_type2,
        'description':description2
    }

    piece_df = pd.DataFrame(data)
    df = pd.concat([df, piece_df], ignore_index=True)
    print(f"{file_name} update successfully")

    return df
```

```python
files = os.listdir("JAVASCRIPT India")
```

```python
DF = pd.DataFrame(columns=['name', 'username', 'location', 'repositories', 'followers', 'language','sorting type','description'])
for file in files:
    df = to_csv_files("JavaScript India", file, "JavaScript")
    DF = pd.concat([DF, df], ignore_index=True)
```
